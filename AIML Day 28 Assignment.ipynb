# Question 1:
Create one array of actual values and another array of predicted values. Compare the two sets
with the confusion matrix.

```python
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix, classification_report, precision_score, recall_score, f1_score
import os
import matplotlib.image as img
import matplotlib.pyplot as plt
```

```python
#array of actual values
ytest = np.array([1,1,1,0,0,0,0,1,1,0,0,1])
ypred = np.array([0,1,1,1,0,1,0,0,1,0,1,1])
```

```python
print(confusion_matrix(ytest,ypred))
```

```python
# in the above confusion matrix, 
TP = 4
FP = 3
FN = 2
TN = 3
```

### Calculation precision, recall and f1 score using formula and simple calculation

```python
precision = TP/(TP+FP)
recall = TP/(TP+FN)
fscore = 2*precision*recall/(precision+recall)
print('precision is {} , recall is {} and f1 score is {}'.format(precision,recall,fscore))
```

### Calculation precision, recall and f1 score using inbuilt function

```python
print('precision is {}, recall is {}, f1_score is {}'.format((precision_score(ytest,ypred)), (recall_score(ytest,ypred)), f1_score(ytest,ypred)))
```

### Calculation precision, recall and f1 score using classification report

```python
print(classification_report(ytest,ypred))
```
