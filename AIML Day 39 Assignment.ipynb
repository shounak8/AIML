```python
'''Importing libraries'''
import pandas as pd
import numpy as np
import os
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

'''importing file-folder'''
os.chdir('C:\\Users\\shoun\\LetsUpgrade\\Datasets')
```

```python
''''''
bb = pd.read_csv('BreadBasket_DMS - Arindam Dev.csv')
bb
```

```python
'''Adding quantity column so if the item is purchased, it will appear as 1'''
bb['Quantity'] = np.nan
bb['Quantity'] = bb['Quantity'].fillna(1)
```

```python
'''Removing the NONE item'''
bb = bb[bb['Item'] != 'NONE']
```

```python
'''Considering relevant variables'''
d1 = bb[['Transaction','Item','Quantity']]
```

```python
'''grouping items as per transactions'''
d2 = d1.groupby(['Transaction','Item'])['Quantity'].sum()
```

```python
d2
```

```python
'''preparing the dataset for applying algorithm'''
d3 = d2.unstack()
```

```python
d4 = d3.applymap(lambda x : 1 if x>0 else 0)
```

```python
d4.head(2)
```

```python
'''Applying Apriori and generating the support dataframe'''
support = apriori(d4, min_support= 0.02, use_colnames=True)
support
```

```python
'''finding association rules'''
association = association_rules(support, metric='lift', min_threshold=1)
```

```python
'''sorting the association result'''
association.sort_values(by = 'lift', ascending = False)
```
